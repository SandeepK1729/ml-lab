```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from seaborn import heatmap
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(60000, 28 * 28)
X_test  = X_test.reshape(10000, 28 * 28)
```

```python
print(f"Train input : {X_train.shape}\tTrain Output : {y_train.shape}")
print(f"Test input  : {X_test.shape} \tTest Output : {y_test.shape}")
```

```python
def print_accuracy_plot_confusion_matrix_for_classification(model,  name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Accuracy with {name} is {accuracy_score(y_test, y_pred)}")
    
    cm = pd.DataFrame(
        confusion_matrix(y_test, y_pred),
        index = [chr(x) for x in range(ord('0'), 10 + ord('0'))],
        columns = [chr(x) for x in range(ord('0'), 10 + ord('0'))],
    )
    
    heatmap(cm, annot = True)
    plt.title("Confusion Matrix of "+ name)
    plt.xlabel("Predicted Values")
    plt.ylabel("Actual Values")
    plt.show()
```

```python
# Naive Bayes based Digit Classification

gnb = GaussianNB()

print_accuracy_plot_confusion_matrix_for_classification(gnb, "Gaussian Naive Bayes")
```

```python
# Suppor Vector Machine Digit Classification

svm = SVC()

print_accuracy_plot_confusion_matrix_for_classification(svm, "Support Vector Machine")
```

```python
# Decision Tree Classifier

dtc = DecisionTreeClassifier()

print_accuracy_plot_confusion_matrix_for_classification(dtc, "Decision Tree Classifier")
```

```python
# K Nearest Neighbours Classifier

knn = KNeighborsClassifier(n_neighbors = 3)

print_accuracy_plot_confusion_matrix_for_classification(knn, "K Nearest Classifier")
```

```python
# Ada Boost Classifier 

abc = AdaBoostClassifier()

print_accuracy_plot_confusion_matrix_for_classification(abc, "Ada Boost Classifier")
```
